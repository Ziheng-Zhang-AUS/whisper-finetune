```python
import whisperx
import torchaudio
import pysubs2

device = "cuda" 
batch_size = 16 # reduce if low on GPU mem
compute_type = "float16" # change to "int8" if low on GPU mem (may reduce accuracy)
```

```python
MP3_PATH = 'data/Schönleber_1.mp3'
```

```python
def srt_to_list_of_dict(srt_path):
    subs = pysubs2.load(srt_path)
    result = []
    for sub in subs:
        d = {'start': sub.start / 1000 , 'end': sub.end / 1000, 'text': sub.text}
        result.append(d)
    return result
```

```python
result = srt_to_list_of_dict('Schönleber_1.srt')
```

```python
# 2. Align whisper output
model_a, metadata = whisperx.load_align_model(language_code='de', device=device)
result = whisperx.align(result, model_a, metadata, MP3_PATH, device, return_char_alignments=False, print_progress=True, combined_progress=True)

print(result) # after alignment
```
